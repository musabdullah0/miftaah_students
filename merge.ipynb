# Try to merge NEO.csv and manual.csv

```python
import requests
import json
import pandas as pd
import time
import numpy as np
import miftaah_utils
```

```python
neo_df = pd.read_csv("NEO.csv")
man_df = pd.read_csv("manual.csv")
```

```python
print("neo_df", neo_df.columns)
print(neo_df.shape)
print("man_df", man_df.columns)
print(man_df.shape)
```

```python
neo_students = {}
group_by_id = neo_df.groupby("student_user_id")

for group_name, group_df in group_by_id:
    neo_id = str(group_df['student_user_id'].iloc[0])
    name = group_df['full_name'].iloc[0]
    email = group_df['email'].iloc[0]
    program = miftaah_utils.find_format(group_df['name'].unique())
    grade = miftaah_utils.find_grade(group_df['joined_at'].iloc[0])
    start_date = miftaah_utils.month_year_format(group_df['joined_at'].iloc[0])
    expected_graduation = miftaah_utils.expected_grad(group_df['joined_at'].iloc[0])
    
    group_df = group_df.replace({np.nan: None})
    periods = group_df[['name', 'class_id', 'enrolled_at',
       'enroll_type', 'percent', 'grade', 'progress', 'started_at',
       'completed_at', 'unenrolled', 'start_at', 'finish_at', 'joined_at']].to_dict('records')
    neo_students[neo_id] = {
        "name": name,
        "email": email,
        "program": program,
        "grade": grade,
        "start_date": start_date,
        "expected_graduation": expected_graduation,
        "periods": periods
    }
```

```python
neo_students
```

```python
class CustomEncoder(json.JSONEncoder):
    def default(self, obj):
        if pd.isna(obj):
            return None
        elif pd.api.types.is_integer_dtype(obj):
            return int(obj)
        else:
            print("idk what type", obj, type(obj))
        return super().default(obj)

json_string = json.dumps(neo_students, cls=CustomEncoder)

with open("neo_data.json", "w+") as f:
    f.write(json_string)
```
